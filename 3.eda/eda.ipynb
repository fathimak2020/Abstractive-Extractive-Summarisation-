```python
import pandas as pd
import glob
import os

from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

### CNA

```python
path = r'../2.text_preprocessing/cna/cna_FINAL.csv' # use your path

cna = pd.read_csv(path)
cna = cna.loc[:,['title','clean_text']]
print(cna.shape)
cna.head()
```

```python
# update stopwords
stop_words = ['say','said','says'] + list(STOPWORDS)
```

```python
#wordcloud for CNA article
cna_longtext = " ".join([one_row for one_row in cna['clean_text']])

wordcloud = WordCloud(stopwords=stop_words, background_color="white", width=1000, height=500).generate(cna_longtext)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("cna_text.png")
```

```python
#wordcloud for CNA article titles
cna_title = " ".join([one_row for one_row in cna['title']])

wordcloud = WordCloud(stopwords=stop_words, background_color="white", width=1000, height=500).generate(cna_title)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("cna_title.png")
```
