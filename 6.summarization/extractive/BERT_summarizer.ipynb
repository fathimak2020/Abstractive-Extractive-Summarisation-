## Import packages

```python
from summarizer import Summarizer
from pprint import pprint
import pandas as pd
pd.set_option("max_colwidth", None)
```

## Read csv with the combined cleaned articles from CNN and CNA

```python
df = pd.read_csv("../../2.text_preprocessing/combined/combined_file.csv")
df.iloc[0]
```

## Load model

```python
model = Summarizer()
```

```python
summarised = []

for i in df["clean_text"]:
    result = model(i, num_sentences=5)
    summarised.append(result)
```

```python
df['bert_extractive'] = summarised
```
